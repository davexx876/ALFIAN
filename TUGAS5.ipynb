```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
from ipywidgets import interact, FloatSlider

# 1. Fungsi dasar filter
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

# 2. Persiapan Data (Sinyal Campuran)
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)
# Sinyal rendah (bersih) + Sinyal tinggi (noise)
signal_clean = np.sin(2 * np.pi * 5 * t)  # 5 Hz
noise = 0.5 * np.sin(2 * np.pi * 150 * t)  # 150 Hz
data_noisy = signal_clean + noise

# 3. Fungsi untuk Update Grafik secara Interaktif
def update_plot(cutoff_freq):
    filtered_data = lowpass_filter(data_noisy, cutoff_freq, fs, order=6)

    # Hitung FFT (untuk melihat spektrum frekuensi)
    fft_noisy = np.abs(np.fft.rfft(data_noisy))
    fft_filtered = np.abs(np.fft.rfft(filtered_data))
    xf = np.fft.rfftfreq(fs, 1/fs)

    plt.figure(figsize=(15, 6))

    # Subplot 1: Sinyal di Domain Waktu
    plt.subplot(1, 2, 1)
    plt.plot(t, data_noisy, label='Berisik (Input)', alpha=0.5, color='gray')
    plt.plot(t, filtered_data, label='Setelah Filter', color='blue', linewidth=2)
    plt.title(f'Domain Waktu (Cutoff: {cutoff_freq} Hz)')
    plt.xlabel('Detik')
    plt.legend()
    plt.grid(True)

    # Subplot 2: Spektrum Frekuensi
    plt.subplot(1, 2, 2)
    plt.plot(xf, fft_noisy, label='Spektrum Asli', color='gray', alpha=0.5)
    plt.plot(xf, fft_filtered, label='Spektrum Terfilter', color='red')
    plt.axvline(cutoff_freq, color='green', linestyle='--', label='Garis Cutoff')
    plt.title('Domain Frekuensi (FFT)')
    plt.xlabel('Frekuensi (Hz)')
    plt.xlim(0, 200) # Fokus ke frekuensi rendah - menengah
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# 4. Membuat Slider Interaktif
interact(update_plot,
         cutoff_freq=FloatSlider(min=1, max=100, step=1, value=20, description='Cutoff (Hz)'))
```
